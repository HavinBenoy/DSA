Insertion Sort Algorithm

```python
# My solution : (best and standard way)

def insertion_sort(arr):
    n = len(arr)
    for i in range(1,n): #The unsorted part of the array
        key = arr[i]
        j = i-1

        while j>=0 and arr[j]> key :
            arr[j+1] = arr[j]
            j -= 1
        
        arr[j+1] = key # This not only satisfies the case in which the arr[j]>key but also correctly places the 'key' element in its own position if arr[j] is not greater than 'key'. 
    return arr


print("Enter the array elements one by one with a space in between",flush = True)
arr = list(map(int,input().split()))
print(f"Array before sorting {arr}")
sorted_arr = insertion_sort(arr)
print(f"Array after sorting {sorted_arr}")

    
```

```python
# Study and understand the working by the above standard method first then only learn this optimized solution.

# Optimized solution : (same as the standard way but here we add a small optimization. Before the while loop we check if the a[j] <= key , if so we can 'continue' the for loop withoy entering the unnecessary while loop)

def insertion_sort(arr):
    n = len(arr)
    for i in range(1,n): #The unsorted part of the array
        key = arr[i]
        j = i-1

        # Skip already sorted elements
        if arr[i - 1] <= key:
            continue
        
        while j>=0 and arr[j]> key :
            arr[j+1] = arr[j]
            j -= 1
        
        arr[j+1] = key # This not only satisfies the case in which the arr[j]>key but also correctly places the 'key' element in its own position if arr[j] is not greater than 'key'. 
    return arr


print("Enter the array elements one by one with a space in between",flush = True)
arr = list(map(int,input().split()))
print(f"Array before sorting {arr}")
sorted_arr = insertion_sort(arr)
print(f"Array after sorting {sorted_arr}")

    
```

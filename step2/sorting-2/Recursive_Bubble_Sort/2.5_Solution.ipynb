```python
# Solution : Most optimal solution .
# Concept of early stopping is used if no swapping is done in an inner looping of the array , meaning the array is sorted.

# --------------------------------------------------
# --------------------------------------------------
def bubble_pass(arr, j, n, swapped):
    if j == n - 1:
        return swapped

    if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
        swapped = True

    return bubble_pass(arr, j + 1, n, swapped)

# --------------------------------------------------
# --------------------------------------------------

def recursive_bubble_sort(arr, n):
    if n == 1:
        return

    swapped = bubble_pass(arr, 0, n, False)

    if not swapped:
        return

    recursive_bubble_sort(arr, n - 1)
# --------------------------------------------------
# --------------------------------------------------

arr = [5, 4, 3, 2, 1]

recursive_bubble_sort(arr, len(arr))

print(arr)
```

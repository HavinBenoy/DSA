Problem : Implement Quick Sort Algorithm

```python
# My solution : 

def pivot(arr,low,high):
    pivot = arr[high] # We use the rightmost element of the array as the pivot element in all cases. Taking this pivot element randomly is another approach which is more optimal.
    i = low -1        # We denote the i such that the  index values from 0 to i are confirmed to be smaller than the pivot element. So we can iterate the array and if we find an element that is lower than the pivot element we can swap it with the (i = i+1) index.  
    for j in range(low,high+1):  # Now we loop thorough all the elements of the array, and if we find a element less than or equal to pivot element , we can swap it with i after incrementing the i
        if arr[j] <= pivot :
            i += 1
            arr[i],arr[j] = arr[j],arr[i]
    return i               # This is the index of the pivot element

def quick_sort(arr,low,high):
    if low>=high: # if the pointers from either sides of the array crosses each other then we simply return , cause it indicates that we iterated through all the elements properly.
        return
    pivot_index = pivot(arr,low,high)
    quick_sort(arr,low,pivot_index-1)
    quick_sort (arr,pivot_index+1, high)

arr = [5,4,2,1,3]
quick_sort(arr,0,len(arr)-1)
print(arr)
```

```python
# Solution : The standard Lomuto Partition Scheme
# This is the exact same as that of my code above but a very small implementation difference.
# In the partition function below , the for loop iterates until the element before the pivot element (high -1 )
# Then after the for loop we seperately increment i and swap the pivot to its position.
# In my implementation this happens automatically in the for loop itself since it iterates to the last element of the array (i.e upto the pivot element). 
# Even though my approach is more easier and cleaner the standard application of the theorem states to divide elements (upto but not including the pivot element) into two positions , then place the pivot to its position.
# I recommend using my solution above itself , but also understands the correct implementation of the Lomuto partition is this solution.

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):   # high excluded
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)

        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)


arr = [5, 4, 2, 1, 3]
quick_sort(arr, 0, len(arr) - 1)
print(arr)
```

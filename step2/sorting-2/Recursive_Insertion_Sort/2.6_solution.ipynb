Problem : Implement Recursive Insertion Sort Algorithm

```python
# My solution:

def iterate_elem(arr,current_index):
    if current_index == len(arr):
        return 
    insertion_sort(arr,current_index)
    iterate_elem(arr,current_index+1)

def insertion_sort(arr,i):

    j = i-1

    if j<0 or arr[j]<= arr[i]:
        return
    
    arr[i] , arr[j] = arr[j] , arr[i]

    insertion_sort(arr,j)


print("Enter a list of number to sort via recursive insertion sort",flush  = True )
arr = list(map(int,input().split()))
print(f"Before sorting : {arr}")
iterate_elem(arr,1)
print(f"After sorting : {arr}")


```

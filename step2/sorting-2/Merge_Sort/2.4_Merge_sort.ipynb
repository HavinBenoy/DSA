Problem : Implement Merge Sort Algorithm

```python
# My solution: 

def merge(arr,low,mid,high):
    left,right = low,mid+1
    temp = []

    while left<=mid and right<=high:
        if arr[left] <= arr[right]:
            temp.append(arr[left])
            left += 1
        else:
            temp.append(arr[right])
            right += 1
    
    while left<=mid:
        temp.append(arr[left])
        left += 1
    
    while right<=high:
        temp.append(arr[right])
        right += 1
    
    k = 0
    for i in range(low,high+1):
        arr[i] = temp [k]
        k += 1
    
def merge_sort(arr,low,high):
    if low >= high:
        return
    mid = (low + high) // 2
    merge_sort(arr,low,mid)
    merge_sort(arr,mid+1,high)
    merge(arr,low,mid,high)

    return arr
    # This return statement is not needed since we are doing in place sorting. Corrected in the optimal approach below

print("Enter a list of number to sort via merge sort",flush  = True )
given_arr = list(map(int,input().split()))
print(f"Before sorting : {given_arr}")
sorted_arr = merge_sort(given_arr , 0 , len(given_arr)-1)
print(f"After sorting : {sorted_arr}")
```

```python
# Most cleaner and python implementation of the merge sort algorithm


def merge(arr, low, mid, high):
    left = low
    right = mid + 1
    temp = []

    while left <= mid and right <= high:
        if arr[left] <= arr[right]:
            temp.append(arr[left])
            left += 1
        else:
            temp.append(arr[right])
            right += 1

    while left <= mid:
        temp.append(arr[left])
        left += 1

    while right <= high:
        temp.append(arr[right])
        right += 1

    arr[low:high+1] = temp     # quicker way to copy a list into another list (even on the desired set of index)


def merge_sort(arr, low, high):
    if low >= high:
        return

    mid = (low + high) // 2

    merge_sort(arr, low, mid)
    merge_sort(arr, mid + 1, high)

    merge(arr, low, mid, high)


arr = list(map(int, input().split()))

print("Before:", arr)

merge_sort(arr, 0, len(arr) - 1)

print("After :", arr)
```

```python
# My solution

class Palindrome_Number:
    def check_pal(self, num):
        rev = 0
        org = num
        while (num >0) :
            last_digit = num % 10
            rev = (rev*10) + last_digit
            num = num // 10
        if rev == org :
            return "Palindrome"
        else:
            return "Not Palindrome"
        
if __name__ == "__main__":
    n = int(input(" Enter a number to check if it is palindrome: "))
    print(n)
    obj = Palindrome_Number() # Creating an object of the class
    result = obj.check_pal(n)
    print(f"The number {n} is {result}") #Function is called with the object created for the class.
         
```

```python
# My Optimal Solution

class Palindrome_Number_Optimal:    #Following a new coding practise of usage of the classes and objects while coding.
    def check_palindrome(self, num):  #Python automatically sends the object itself to the method so the method knows which object it is working on. So "self" should be the first argument in defining the functions within a class.
        value = str(num)
        reverse  = value[::-1]
        if value == reverse :
            return "Palindrome"
        else:
            return "Not Palindrome"

if __name__ == "__main__":
    n = int(input(" Enter a number to check if it is palindrome: "))
    print(n)
    obj = Palindrome_Number_Optimal() # Creating an object of the class
    result = obj.check_palindrome(n)
    print(f"The number {n} is {result}") #Function is called with the object created for the class.
```

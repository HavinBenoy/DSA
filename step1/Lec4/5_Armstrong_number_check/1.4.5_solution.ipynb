# Problem : Check if a number is Armstrong Number or not

```python
class armstrong:
    def isarmstrong(self, str_num):
        sum = 0
        length = len(str_num)
        for i in str_num:
            sum = sum + (int(i)**length)
        print(sum)
        if sum == int(str_num):
            return 1
        else :
            return 0

if __name__ == "__main__":
    num = input("Enter a number to check if its armstrong: ")
    obj = armstrong()
    if obj.isarmstrong(num) == 1:
        print("Given number is armstrong")
    else:
        print("Given number is not armstrong")
```

Another similar code without changing to str is given below:

```python
# Define a class to check Armstrong numbers
class ArmstrongChecker:

    @staticmethod
    def is_armstrong(num):
        k = len(str(num))  # Number of digits
        sum = 0
        n = num

        while n > 0:
            ld = n % 10             # Last digit
            sum += ld ** k          # Add ld^k
            n = n // 10             # Remove digit

        return sum == num

# Test the class method
number = 153

if ArmstrongChecker.is_armstrong(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

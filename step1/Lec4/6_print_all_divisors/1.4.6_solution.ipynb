Problem : Print all Divisors of a given Number

1.BruteForce Approach

```python
class divisors:
    def get_divisors(self, num):
        res = []
        for i in range(1,num+1):
            if num % i == 0 :
                res.append(i)
        return res
    
if __name__ == "__main__":
    n = int(input("Enter a number to find the divisors"))
    obj = divisors()
    divisor_list = obj.get_divisors(n)
    print("Divisors of the given number are: \n",divisor_list)


```

2.optimal approach:



We can optimise the previous approach by using the property that for any non-negative integer n, if d is a divisor of n then n/d is also a divisor of n. This property is symmetric about the square root of N. Thus, by traversing just the first half we can avoid redundant iteration and computations improving the efficiency of the algorithm.


Iterate from 1 to sqrt(N) and for every divisor found, if N/divisor is distinct, add that to the list of divisors as well.

```python
# my optimal solution

import math

class divisors:
    def get_divisors(self, num):
        res = []
        for i in range(1,int(math.sqrt(num))+1):
            if num % i == 0 :
                res.append(i)
                # other than the perfect sqaure case , in each of the num=a*b scenario both a and b are the divisors of the num while viewing the entire set of divisors . Perfect sqaure case if avoided to avoid redundency since in that case num = a*a so we dont have to add 'a' twice. 
                if num//i != i :
                    res.append(num//i)
        return res
    
if __name__ == "__main__":
    n = int(input("Enter a number to find the divisors"))
    obj = divisors()
    divisor_list = obj.get_divisors(n)
    print("Divisors of the given number are: \n",divisor_list)


```

```python
# Most optimal solution (Same logic as my optimal solution above but structured and commented):

import math

class Solution:
    # Function to get all divisors
    def getDivisors(self, N):
        # Create list to store divisors
        res = []

        # Loop from 1 to square root of N
        for i in range(1, int(math.isqrt(N)) + 1):
            # Check if i divides N
            if N % i == 0:
                # Add i to result
                res.append(i)

                # If N // i is not the same, add that too
                if i != N // i:
                    res.append(N // i)

        # Return the list of divisors
        return res

# Create object of Solution class
sol = Solution()

# Input number
N = 36

# Get divisors
result = sol.getDivisors(N)

# Print the result
print("Divisors of", N, ":", *result)
```

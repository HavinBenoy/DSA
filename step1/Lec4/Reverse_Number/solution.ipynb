```python
#My solution

def reverse(num):
    rev = 0
    while (num >0) :
        last_digit = num % 10
        rev = (rev*10) + last_digit
        num = num // 10
    return rev

if __name__ == "__main__":
    n = int(input("Please enter a number to reverse: "))
    print(f"Reverse of {n} is {reverse(n)}")
```



Time Complexity: O(log₁₀N),The number of iterations in the loop depends on the number of digits in the number N. Since each digit is processed once using constant time operations (modulo, division, multiplication, and addition), the total time taken is proportional to the number of digits, which is log₁₀N.

```python
# My optimal solution approach

def reverse_optimal(num):
    number = str(num)
    reverse = number[::-1]
    return int(reverse)



if __name__ == "__main__":
    n = int(input("Please enter a number to reverse: "))
    print(f"Reverse of {n} is {reverse_optimal(n)}") 
```

Now the time complexity is O(1)

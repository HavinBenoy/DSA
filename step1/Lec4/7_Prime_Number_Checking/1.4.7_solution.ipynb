# Problem : Check if a number is prime or not

My Solution 1 : Bruteforce approach 

```python
class prime():
    def is_prime(self, num):
        count = 0
        for i in range(2, num): #Checking all the other numbers between 1 and the number itself , to look for any other divisors.
            if num % i == 0:
                count += 1
        if count == 0:
            return True
        else :
            return False

if __name__ == "__main__":
    n = int(input("Enter a number to check whether it is a prime number: "))
    if n<=1:
        print("Prime number are not defined for numbers less than 2")
    else:
        obj = prime()
        if obj.is_prime(n):
            print(f"{n} is a prime number")
        else :
            print(f"{n} is not a prime number")
```

My solution 2 : Optimal approach

logic : We can optimise the algorithm by only iterating up to the square root of n when checking for factors. This is because if n has a factor greater than its square root, it must also have a factor smaller than its square root.

This property is symmetric about the square root of n by traversing just the first half we can avoid redundant iteration and computations improving the efficiency of the algorithm.

```python
import math

class prime():
    def is_prime(self, num):
        count = 0
        for i in range(2, int(math.sqrt(num)+1)): #Checking all the other numbers between 1 and the number itself , to look for any other divisors.
            if num % i == 0:
                count += 1
        if count == 0:
            return True
        else :
            return False

if __name__ == "__main__":
    n = int(input("Enter a number to check whether it is a prime number: "))
    if n<=1:
        print("Prime number are not defined for numbers less than 2")
    else:
        obj = prime()
        if obj.is_prime(n):
            print(f"{n} is a prime number")
        else :
            print(f"{n} is not a prime number")
```

Solution 3 : 
    Most optimal solution

logic :
    Instead of checking:

    2, 3, 4, 5, 6, 7...

    it checks:

    only odd numbers after 2
    because even numbers can never divide an odd prime

    So for large numbers, it reduces iterations nearly by half.

```python
import math

def is_prime(num):
    if num <= 1:
        return False

    if num == 2:
        return True

    if num % 2 == 0:
        return False

    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False

    return True


n = int(input("Enter a number: "))

if is_prime(n):
    print(f"{n} is a prime number")
else:
    print(f"{n} is not a prime number")
```

Problem : Factorial of a Number : Iterative and Recursive

```python
# My solution : Iterative and Recursive approaches

def factorial_iterative(n):  # Best approach since it has Time complexity (hereby will be represented as TC) of O(n) and space complexity (hereby will be represented as SC) of O(1)
    fact = 1
    for i in range (1,n+1):
        fact *= i
    return fact


def factorial_recursion(n): # TC= O(n)  & SC = O(n)
    if n == 1:
        return 1
    return n * factorial_recursion(n-1)


k = int(input("Enter a number: "))

fact_iteration = factorial_iterative(k)

fact_recursion = factorial_recursion(k)

print(f"Factorial using iteration = {fact_iteration}\nFactorial using Recursion = {fact_recursion}")
```

Problem : Print N to 1 using Recursion

```python
# My solution: (Best)

def n_to_one(n):
    if n == 0 :
        return
    print(n ,end = " ")
    n_to_one(n-1)


n = int(input("Enter a number"))
n_to_one(n)
```

Solution 1 : Forward tracking for N-to-1 (reverse of forward tracking in 1 to N) (Print first then recursion call)

```python
def print_n_to_1_forward(n):
    if n == 0: return
    print(n, end = " ")                      # Print immediately
    print_n_to_1_forward(n - 1)  # Move to next state


n = int(input("Enter a number"))
print_n_to_1_forward(n)
```

Solution 2 : Backtracking for N-to-1 (Recursion first then print statement ) (Have index value to track)

```python
def print_n_to_1_backward(n, i=1):
    if i > n: return
    print_n_to_1_backward(n, i + 1) # Drill down first
    print(i, end =  " ")                          # Print on the way up


n = int(input("Enter a number"))
print_n_to_1_backward(n)
```

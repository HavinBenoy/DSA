Problem: Sum of first N Natural Numbers

```python
# My solution: Brute Force Approach, Equation based Approach and Recusion Approach

def n_sum_Bruteforce(n):
    total = 0
    for i in range (n+1):
        total += i
    return total

def n_sum_equation(n):          # Most optimial and best approach with a complexity of O(1) in time and space of execution. 
    total = (n*(n+1)) // 2
    return total

def n_sum_recursion(n):
    if n<1:
        return 0
    return n + n_sum_recursion (n-1)

k = int(input("Enter a number : "))

sum_Bruteforce = n_sum_Bruteforce(k)

sum_Equation = n_sum_equation(k)

sum_Recusion = n_sum_recursion(k)

print(f"Sum via Bruteforce appraoch = {sum_Bruteforce} \nSum via Equation approach = {sum_Equation} \nSum via Recursion appraoch = {sum_Recusion} ")

```

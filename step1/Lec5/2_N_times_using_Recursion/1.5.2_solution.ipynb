# Problem : Print Name N times using Recursion

my solution 1 : Straight forward approach

```python
def printntimes(n):
    if n != 1 :
        printntimes(n-1)
    print("DSA \n")
```

```python
n = int(input("Enter a number: "))
printntimes(n)
```

Note : This approach correct but have certain drawbacks. The main problem is the lack of a good stop condition. if the input n = 0 then funtion might continue call itself recursively until the stack overflow occurs . so follow the below best appraoch of defining a good stopping condition.

Solution 2 : Optimal solution (More like a standard approach to the problem)

```python
def print_n_times(n):
    
    # Base Case:
    # When n becomes 0, stop recursion
    if n == 0:
        return
    
    # Print the name once
    print("DSA")
    
    # Recursive call for remaining times
    print_n_times(n - 1)


# Input
n = int(input("Enter a number: "))

# Function call
print_n_times(n)
```

Problem : fibonacci Series up to Nth term

```python
# My solution : Brute Force approach 

def fib(n):
    result = []
    result.append(0)
    if n==0:
        return result
    result.append(1)
    if n==1:
        return result
    for i in range(2,n+1):   # 0 based index is preffered in the question
        result.append(result[i-1]+result[i-2])
    return result

n = int(input("Enter the size of the fibanocci series: "))      
fib_series = fib(n)
print(f"Fibanocci series of size {n} is :\n{fib_series}")
```

```python
# Solution : Optimal Approach
# For calculating the ith term we only need the last and second last term i.e (i-1)th and (i-2)th term, so we don't need to maintain the whole array

def main():
    n = 5

    # If n is 0, only the first term is printed
    if n == 0:
        print(f"The Fibonacci Series up to {n}th term:")
        print(0)
    else:
        second_last = 0  # (i-2)th term
        last = 1         # (i-1)th term

        print(f"The Fibonacci Series up to {n}th term:")
        print(f"{second_last} {last}", end=" ")

        for i in range(2, n + 1):
            cur = last + second_last  # Current ith Fibonacci number
            second_last = last        # Move window
            last = cur
            print(cur, end=" ")

if __name__ == "__main__":
    main()
```

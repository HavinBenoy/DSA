Problem : Find the highest & lowest frequency element in an array of elements

```python
# My solution: not optimal (two for loops to find high and low values seperately is not necessary)
# But better than the Bruteforce
# TC = O(3n) i.e aprox.equal to  O(n)

from collections import defaultdict

def high_n_low(arr):
    dict = defaultdict(int)

    for i in arr:
        dict[i] += 1     # The case of dict[i] not being initialized will be handled automatically by the defaultdict by setting a default value 0 to that dict[i] automatically and then proceeding with the increment operation defined here.

    high_val = max(dict.values())
    low_val = min(dict.values())

    high = [k for k,v in dict.items() if v==high_val]
    low = [k for k,v in dict.items() if v==low_val]
    
    return high , low

k = print("Enter a set of numbers to find their frequency", flush= True)
k  = list(map(int, input().split()))
print (k)
high, low  = high_n_low (k)
print(f"Highest frequent element in the given list is {high}\nLowest frequent element in the given list is {low}")

```

```python
# Solution : Bruteforce
# TC = O(n^2)

class FrequencyCounter:
    def countFreq(self, arr):
        n = len(arr)                     # Get array length
        visited = [False] * n           # Track if an element has already been counted
        maxFreq = 0                     # Store max frequency seen so far
        minFreq = n                     # Store min frequency (start high)
        maxEle = 0                      # Element with max frequency
        minEle = 0                      # Element with min frequency

        for i in range(n):

            # Skip already processed elements
            if visited[i]:
                continue

            # Count frequency of arr[i]
            count = 1
            for j in range(i + 1, n):
                if arr[i] == arr[j]:
                    visited[j] = True   # Mark as visited
                    count += 1

            # Update max frequency info
            if count > maxFreq:
                maxEle = arr[i]
                maxFreq = count

            # Update min frequency info
            if count < minFreq:
                minEle = arr[i]
                minFreq = count

        # Print results
        print("The highest frequency element is:", maxEle)
        print("The lowest frequency element is:", minEle)


if __name__ == "__main__":
    fc = FrequencyCounter()                     # Create object
    arr = [10, 5, 10, 15, 10, 5]                # Sample input
    fc.countFreq(arr)                           # Run frequency analysis
```

```python
# Solution : Optimal (best) 
# TC = O(2n) i.e aprox.equal to  O(n)

class FrequencyCounter:
    def Frequency(self, arr):
        freq_map = {}                # Dictionary to store frequency of each element

        # Count frequencies
        for num in arr:
            freq_map[num] = freq_map.get(num, 0) + 1   #impo : Else we can use a defualtdict as in my solution aboue to handle these initialization conditions.

        maxFreq = 0
        minFreq = len(arr)
        maxEle = 0
        minEle = 0

        # Iterate through dictionary to find max and min frequency elements
        for element, count in freq_map.items():
            if count > maxFreq:
                maxFreq = count
                maxEle = element

            if count < minFreq:
                minFreq = count
                minEle = element

        # Print results
        print("The highest frequency element is:", maxEle)
        print("The lowest frequency element is:", minEle)


if __name__ == "__main__":
    fc = FrequencyCounter()
    arr = [10, 5, 10, 15, 10, 5]
    fc.Frequency(arr)
```

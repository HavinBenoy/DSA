Problem : Binary Search: Explained


Problem statement: You are given a sorted array of integers and a target, your task is to search for the target in the given array. Assume the given array does not contain any duplicate numbers.

```python
# My solution : iterative solution

def binary_search(arr,k):
    low = 0
    high = len(arr)-1
    while low<=high: 
        mid = (low+high) // 2
        if arr[mid]==k:
            print(f"Element {k} is found on the index {mid}.")
            return
        elif(k < arr[mid]):
            high = mid-1
        else:
            low = mid+1
    print(f"The element {k} doen not exist in the array")
    return

arr = [1,2,3,4,5,6,7,8,9,18]
k = 6
binary_search(arr,k)
```

```python
# My solution : recursive solution

def binary_search_recursive(arr,high,low,k):
    if high<low:
        print(f"Element {k} is not found")
        return
    mid = (low+high)//2
    if arr[mid] == k:
        print(f"Element {k} is found on index {mid}")
        return
    elif (k<arr[mid]):
        binary_search_recursive(arr,mid-1,low,k)
    else:
        binary_search_recursive(arr,high,mid+1,k)



arr = [1,2,3,4,5,6,7,8,9,18]
k = 6
low = 0
high = len(arr)-1
binary_search_recursive(arr,high,low,k)
```

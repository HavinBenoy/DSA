Problem : Floor and Ceil in Sorted Array

Problem Statement: ou're given an sorted array arr of n integers and an integer x. Find the floor and ceiling of x in arr[0..n-1]. The floor of x is the largest element in the array which is smaller than or equal to x. The ceiling of x is the smallest element in the array greater than or equal to x

Examples

Example 1:
Input Format: n = 6, arr[] ={3, 4, 4, 7, 8, 10}, x= 5
Result: 4 7
Explanation: The floor of 5 in the array is 4, and the ceiling of 5 in the array is 7.

Example 2:
Input Format: n = 6, arr[] ={3, 4, 4, 7, 8, 10}, x= 8
Result: 8 8
Explanation: The floor of 8 in the array is 8, and the ceiling of 8 in the array is also 8

```python
# My solution:

def find_floor_and_ceil(arr,target):
    low = 0
    high = len(arr)-1

    if target > arr[high]:
        return [high,-1]
    elif target < arr[low]:
        return [-1,low]

    while(low<high):
        mid = (low+high)//2
        if arr[mid]==target:
            return [mid,mid]
        elif high - 1 == low:
            return [low,high]
        elif (arr[mid]<target):
            low = mid
        else:
            high = mid

arr = [3, 4, 4, 7, 8, 10,11,13]
tar = 22
floor_value, ceil_value = find_floor_and_ceil(arr,tar)
if floor_value != -1  and ceil_value != -1:
    print(f"floor value is {arr[floor_value]} and ceil value is {arr[ceil_value]}") 
elif floor_value != -1:
    print(f"floor value is {arr[floor_value]} and ceil value doesnt exist")
else:
    print(f"ceil value is {arr[ceil_value]} and the floor value doesnt exist ")
```

```python
# Solution : (My solution is more optimal so follow it)

class FloorCeilFinder:
    def find_floor(self, arr, x):
        low, high = 0, len(arr) - 1
        ans = -1
        while low <= high:
            mid = (low + high) // 2
            if arr[mid] <= x:
                ans = arr[mid]  # Potential floor
                low = mid + 1
            else:
                high = mid - 1
        return ans

    def find_ceil(self, arr, x):
        low, high = 0, len(arr) - 1
        ans = -1
        while low <= high:
            mid = (low + high) // 2
            if arr[mid] >= x:
                ans = arr[mid]  # Potential ceil
                high = mid - 1
            else:
                low = mid + 1
        return ans

    def get_floor_and_ceil(self, arr, x):
        floor = self.find_floor(arr, x)
        ceil = self.find_ceil(arr, x)
        return floor, ceil

# Usage
arr = [3, 4, 4, 7, 8, 10,11,13]
x = 22
finder = FloorCeilFinder()
f, c = finder.get_floor_and_ceil(arr, x)
print(f"The floor and ceil are: {f} {c}")

```

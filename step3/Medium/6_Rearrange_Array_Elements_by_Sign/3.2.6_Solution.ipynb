#Problem: Rearrange Array Elements by Sign

Problem Statement: There’s an array ‘A’ of size ‘N’ with an equal number of positive and negative elements. Without altering the relative order of positive and negative elements, you must return an array of alternately positive and negative values.

Please check the example for a better understanding of the problem.

```python
# My solution : Bruteforce Approach

def sign_altered_array(arr):
    pos = []
    neg = []
    res = []
    for i in range(len(arr)):
        if arr[i]>0 :
            pos.append(arr[i])
        else:
            neg.append(arr[i])
    i,j,k = 0,0,0
    while i!=len(arr):
        if i % 2 == 0:
            res.append(pos[j])
            j += 1
        else:
            res.append(neg[k])
            k += 1
        i += 1
    return res

arr = [1,2,-3,-1,-2,3]
resultant_array = sign_altered_array(arr)
print(f"Given array : {arr}\nSign altered array : {resultant_array}")
```

```python
# Solution : Better implementation of the bruteforce approach.
# This is same my bruteforce solution above but the second loop is implemented in a more clean and optimal way.

class ArrayManipulator:
    def rearrange_by_sign(self, A, n):
        pos = []  # List to hold positive numbers
        neg = []  # List to hold negative numbers

        # Step 1: Separate positives and negatives
        for i in range(n):
            if A[i] > 0:
                pos.append(A[i])
            else:
                neg.append(A[i])

        # Step 2: Place positives at even indices, negatives at odd indices
        for i in range(n // 2):
            A[2 * i] = pos[i]      # Even index → positive
            A[2 * i + 1] = neg[i]  # Odd index → negative

        return A

# Driver code
if __name__ == "__main__":
    A = [1, 2, -4, -5]
    n = len(A)

    obj = ArrayManipulator()
    result = obj.rearrange_by_sign(A, n)

    print(*result)  # Print result with space
```

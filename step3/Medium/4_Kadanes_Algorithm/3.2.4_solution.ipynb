Problem : Kadane's Algorithm : Maximum Subarray Sum in an Array

Problem Statement: Given an integer array nums, find the subarray with the largest sum and return the sum of the elements present in that subarray.

A subarray is a contiguous non-empty sequence of elements within an array.

```python
# My solution :

def max_subarray(arr):
    max_sum = float('-inf')
    current_sum = 0
    n = len(arr)
    sum_record = []# we dont need a dictionary since only the sum of the max subarray is asked in the question , not to print the subarray.
    sum_record.append(0) #so that we can compare the sum upto n elements also , rather than just starting from h
    for i in range(n):
        current_sum += arr[i]
        
        for j in sum_record:
            if current_sum-j > max_sum:
                max_sum = current_sum-j

        sum_record.append(current_sum)
    return max_sum

arr = [2, 3, 5, -2, 7, -4]
print(f"maximum sum of subarray is: {max_subarray(arr)}")

```

```python
# Solution : Better approach . It is the same logic as that of my solution .
# But instead of checking the difference between each and every sums in the list of sumrecords , this solution keeps the minimum sum value of the array at every point .i.e the minimum prefix sum.
# so that only the difference between current sum and the most minimum prefix sum is needed to be checked rather than the entire sums.

def max_subarray(arr):
    prefix_sum = 0
    min_prefix = 0
    max_sum = float('-inf')

    for num in arr:
        prefix_sum += num

        max_sum = max(max_sum, prefix_sum - min_prefix)

        min_prefix = min(min_prefix, prefix_sum)

    return max_sum

arr = [2, 3, 5, -2, 7, -4]
print(f"maximum sum of subarray is: {max_subarray(arr)}")
```

```python
# Solution : Kadanes solution : Most optimal solution to the problem :

from typing import List

class Solution:
    # Function to find maximum sum of subarrays
    def maxSubArray(self, nums: List[int]) -> int:
        
        # maximum sum
        maxi = float('-inf') 
        
        # current sum of subarray
        sum = 0 
        
        # Iterate through the array
        for i in range(len(nums)):
            
            # Add current element to the sum
            sum += nums[i] 
            
            # Update maxi if current sum is greater
            if sum > maxi:
                maxi = sum 
            
            # Reset sum to 0 if it becomes negative
            if sum < 0:
                sum = 0 
        
        # Return the maximum subarray sum found
        return maxi

if __name__ == "__main__":
    arr = [ -2, 1, -3, 4, -1, 2, 1, -5, 4 ]

    # Create an instance of Solution class
    sol = Solution()

    maxSum = sol.maxSubArray(arr)

    # Print the max subarray sum
    print(f"The maximum subarray sum is: {maxSum}")
```

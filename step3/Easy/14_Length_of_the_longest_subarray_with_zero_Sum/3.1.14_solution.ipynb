Problem : Length of the longest subarray with zero Sum

In the previous problem the optimal solution could only handle arrays with positive numbers . But this problem requires the ability to deal with both positive and negative numbers. so this is going to be the best solution to the subarrays problem than the last one.

```python
# My solution : Bruteforce and non optimal.


def find_long_array(arr,arr_size):
    sum_elem = 0
    n = len(arr)
    if arr_size<1:
        print("0")
        return
    else:
        for i in range(0,n-arr_size+1): # starting index
            sum = 0
            for j in range (i,i+arr_size):
                sum += arr[j]
            if sum==sum_elem:
                print(f"length of the longest subarray is : {arr_size}")
                return
        
        find_long_array(arr,arr_size-1)
            
arr = [6, -2, 2, -8, 1, 7, 4, -10]
n = len(arr)
find_long_array(arr,n)

```

```python
# Solution : Bruteforce 

# compute length of the longest subarray with sum 0
def solve(a: list[int]) -> int:
    # store best length found so far
    max_len = 0
    # map prefix sum -> first index seen
    sum_index = {}
    # running prefix sum
    s = 0

    # iterate through the array
    for i, val in enumerate(a):
        # update running sum
        s += val

        # if sum is zero, subarray [0..i] has zero sum
        if s == 0:
            # update best length
            max_len = i + 1
        # if this sum seen before, subarray (prevIndex..i] has zero sum
        elif s in sum_index:
            # maximize length using previous index
            max_len = max(max_len, i - sum_index[s])
        # first time seeing this sum, store its index
        else:
            sum_index[s] = i

    # return best length
    return max_len

# program entry
def main():
    # sample input
    a = [9, -3, 3, -1, 6, -5]
    # print result
    print(solve(a))

# run main
if __name__ == "__main__":
    main()
```

```python
# Solution : Most optimal approach for subarrays . Considers both positive and negative numbers within the subarray.


# compute length of the longest subarray with sum 0
def maxLen(A: list[int], n: int) -> int:
    # map prefix sum -> first index seen
    mpp: dict[int, int] = {}
    # best length so far
    maxi = 0
    # running prefix sum
    s = 0

    # iterate over the array
    for i in range(n):
        # update running sum
        s += A[i]

        # if sum is zero, subarray [0..i] has zero sum
        if s == 0:
            # update best length
            maxi = i + 1
        # otherwise check if this sum was seen before
        else:
            # when seen, zero-sum segment between previous index + 1 and i
            if s in mpp:
                # maximize length
                maxi = max(maxi, i - mpp[s])
            # first time seeing this sum
            else:
                # record index
                mpp[s] = i

    # return best length
    return maxi

# program entry
def main():
    # sample input
    A = [9, -3, 3, -1, 6, -5]
    # compute size
    n = len(A)
    # print result
    print(maxLen(A, n))

# run main
if __name__ == "__main__":
    main()
```

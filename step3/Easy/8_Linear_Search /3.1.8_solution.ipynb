Problem : Linear Search 

```python
# My solution :

def linear_search(arr,k):
    index = -1
    for ind,val in enumerate(arr):
        if val == k:
            index = ind
            break
    return index

arr = [1,2,3,4,5,6,7,8,9]
search_element = 7
res = linear_search(arr,search_element)
if res == -1:
    print(f"The element is not found in the array")
else:
    print(f"The element is found in the index : {res}")
```

Problem : Count Maximum Consecutive One's in the array

```python
# My solution :
def count_max_ones(arr):
    flag = False
    max_count = 0
    count = 0

    for i in arr:
        if i==1 :
            flag = True
            count += 1

        elif i==0 and flag == True:
            if max_count<count:
                max_count = count
                count = 0
            flag = False

    if max_count<count:
        max_count = count

    return max_count

arr = [1, 0, 1, 1, 0, 1]
max_ones = count_max_ones(arr)
print(f"maximum number of consecutive ones is : {max_ones}")


        


        
```

```python
# Solution: More simpler and easier and optimal implementation.

class Solution:
    # Function to find maximum consecutive 1's in an array
    def findMaxConsecutiveOnes(self, nums):
        # Variable to store current count of consecutive 1's
        cnt = 0
        # Variable to store maximum consecutive 1's
        maxi = 0

        # Traverse the array
        for i in range(len(nums)):
            # If current element is 1, increment count
            if nums[i] == 1:
                cnt += 1
            else:
                # If element is 0, reset count
                cnt = 0

            # Update maximum if current count is greater
            maxi = max(maxi, cnt)

        # Return maximum consecutive 1's
        return maxi

# Input array
nums = [1, 1, 0, 1, 1, 1]

# Create Solution object
obj = Solution()

# Get answer
ans = obj.findMaxConsecutiveOnes(nums)

# Print result
print("The maximum consecutive 1's are", ans)
```

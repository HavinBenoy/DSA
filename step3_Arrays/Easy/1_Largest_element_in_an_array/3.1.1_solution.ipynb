Problem : Find the Largest element in an array

```python
# My solution 

def largest_element_iteration(arr): # Most standard and optimal appraoch 
    max_element = arr[0]
    for i in arr:
        if i > max_element :
            max_element = i
    return max_element

def largest_element_by_func (arr):      # Easiest Approach
    largest_element = max(arr)
    return largest_element

k =[8, 10, 5, 7, 9]
max_num = largest_element_iteration(k)
lar_num = largest_element_by_func(k)
print(f"Largest number is {max_num}")

print(f"Largest number is {lar_num}")
```

Problem : Longest Subarray with given Sum K(Positives).

Problem Statement: Given an array nums of size n and an integer k, find the length of the longest sub-array that sums to k. If no such sub-array exists, return 0.

Examples
Example 1:
Input:
 nums = [10, 5, 2, 7, 1, 9], k = 15  
Output:
 4  
Explanation:
 The longest sub-array with a sum equal to 15 is [5, 2, 7, 1], which has a length of 4. This sub-array starts at index 1 and ends at index 4, and the sum of its elements (5 + 2 + 7 + 1) equals 15. Therefore, the length of this sub-array is 4.

Example 2:
Input:
 nums = [-3, 2, 1], k = 6  
Output:
 0  
Explanation:
 There is no sub-array in the array that sums to 6. Therefore, the output is 0.


```python
# My solution:

def find_long_array(arr,sum_elem,arr_size):
    n = len(arr)
    if arr_size<1:
        print("0")
        return
    else:
        for i in range(0,n-arr_size+1): # starting index
            sum = 0
            for j in range (i,i+arr_size):
                sum += arr[j]
            if sum==sum_elem:
                print(f"length of the longest subarray is : {arr_size}")
                return
        
        find_long_array(arr,sum_elem,arr_size-1)
            
arr = [-1, 1, 1]
sum_element = 1
n = len(arr)
find_long_array(arr,sum_element,n)
```

```python
# Solution : Optimal approach but this solution will not work with negative numbers in the array. If the negative numbers also needed try one of the  above solutions.

class Solution:
    # Function to find the length of longest subarray having sum k
    def longestSubarray(self, nums, k):
        n = len(nums)
        
        # To store the maximum length of the subarray
        maxLen = 0
        
        # Pointers to mark the start and end of window
        left = 0
        right = 0
        
        # To store the sum of elements in the window
        sum = nums[0]
        
        # Traverse all the elements
        while right < n:
            
            # If the sum exceeds K, shrink the window
            while sum > k and left <= right:
                sum -= nums[left]
                left += 1
            
            # Store the maximum length
            if sum == k:
                maxLen = max(maxLen, right - left + 1)
            
            right += 1
            if right < n:
                sum += nums[right]
        
        return maxLen


nums = [10, 5, 2, 7, 1, 9]
k = 15

# Creating an object of Solution class
sol = Solution()

# Function call to find the length
# of longest subarray having sum k
ans = sol.longestSubarray(nums, k)

print(f"The length of longest subarray having sum k is: {ans}")


```

# Best and most optimal solution for this problem of subarray is given below . This solution can handle both positive and negative numbers within the array.

#Core Idea#

Instead of checking every possible subarray (which would take O(n²)), we use:

Prefix Sum: 
prefix_sum[i] = sum of elements from index 0 to i


HashMap (first_occurrence): 
Stores the first index where a particular prefix sum appears.

Mathematical Insight

Suppose:


Prefix sum at index i = curr_sum

We want a subarray ending at i whose sum is k.


If some earlier prefix sum was:


curr_sum−k


then:


curr_sum−(curr_sum−k)=k


Meaning the elements between those two indices sum to k.


Example

Array: [2, 3, 1, 2]

k = 4


Prefix sums:


Index:      0   1   2   3
Element:    2   3   1   2
Prefix:     2   5   6   8


At index 3:


prefix_sum = 8

remaining = 8 - 4 = 4


If we had seen prefix sum 4 earlier, then the subarray after that index up to index 3 would sum to 4.

```python
# Solution : Most optimal solution (which considered all the cases regardless of whether the sign of the nnumbers are positive or negative)

def longest_subarray(arr, k):
    prefix_sum = 0
    max_len = 0

    first_occurrence = {}

    for i in range(len(arr)):
        prefix_sum += arr[i]

        # Entire array from 0 to i
        if prefix_sum == k:
            max_len = i + 1

        # Store first occurrence only
        if prefix_sum not in first_occurrence:
            first_occurrence[prefix_sum] = i

        remaining = prefix_sum - k

        if remaining in first_occurrence:
            length = i - first_occurrence[remaining]
            max_len = max(max_len, length)

    return max_len

arr = [-1, 1, 1]
sum_element = 1
print(longest_subarray(arr,sum_element))
```

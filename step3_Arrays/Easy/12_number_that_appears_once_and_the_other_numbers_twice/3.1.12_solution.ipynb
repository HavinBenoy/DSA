Problem : Find the number that appears once, and the other numbers twice

Example 1:
Input Format: arr[] = {2,2,1}
Result: 1
Explanation: In this array, only the element 1 appear once and so it is the answer.


Example 2:
Input Format: arr[] = {4,1,2,1,2}
Result: 4
Explanation: In this array, only element 4 appear once and the other elements appear twice. So, 4 is the answer.

```python
# My Solution:

def number_that_appear_ones(arr):
    arr.sort()
    n = len(arr)
    i = 0
    while i<n:
        if i==len(arr)-1:
            print(arr[i])
            break
        elif arr[i] == arr[i+1]:
            i += 2
        else:       # Since all the other elements are guaranteed to be appearing twice.
            print(arr[i])
            break

arr = [4,1,2,1,2]
number_that_appear_ones(arr)    
```

```python
# Solution : Bruteforce approach : It is less optimal than my solution.
# This is the most straigh forward and direct way

class Solution:
    # Function to find the element that appears only once
    def getSingleElement(self, arr):
        n = len(arr)

        # Check each element's count
        for i in range(n):
            num = arr[i]
            count = 0

            # Count how many times num appears
            for j in range(n):
                if arr[j] == num:
                    count += 1

            # Return if it appears exactly once
            if count == 1:
                return num

        return -1  # fallback

# Driver code
arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```

```python
#Solution : Bruteforce approach 2
# It is also less optimal than my solution because if we do have values like 1 million in the input array even if the size of the array might be less.
# in such cases, we need that much (1million ) sized array to store the values. so its not scalable solution.

class Solution:
    # Function to find the element that appears only once using hashing
    def getSingleElement(self, arr):
        n = len(arr)

        # Step 1: Find max value to create hash array
        maxi = max(arr)

        # Step 2: Create and initialize hash list
        hash_arr = [0] * (maxi + 1)

        # Step 3: Count frequency of each number
        for num in arr:
            hash_arr[num] += 1

        # Step 4: Find and return the number that appears once
        for num in arr:
            if hash_arr[num] == 1:
                return num

        return -1  # fallback

# Driver code
arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```

```python
# Solution : we have already discussed that if we xor two same values it will end up as zero.
# so if we xor all the elements then we will get the unique number as a result.

class Solution:
    # Function to find the single non-repeating element using XOR
    def getSingleElement(self, arr):
        xorr = 0

        # XOR all elements — duplicates cancel out
        for num in arr:
            xorr ^= num

        return xorr

# Driver code
arr = [4, 1, 2, 1, 2]
obj = Solution()
ans = obj.getSingleElement(arr)
print("The single element is:", ans)
```

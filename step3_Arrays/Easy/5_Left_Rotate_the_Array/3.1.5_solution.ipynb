Problem :Left Rotate the Array by One

```python
# My solution: Optimal and easy solution

def left_rotate(arr):
    n = len(arr)
    temp = arr[0]
    for i in range(0,n-1):
        arr[i] = arr[i+1]
    arr[n-1] = temp
    return arr

arr = [-1, 0, 3, 6]
print(f"Before: {arr}")
rotated_arr = left_rotate(arr)
print(f"After : {rotated_arr}")
```

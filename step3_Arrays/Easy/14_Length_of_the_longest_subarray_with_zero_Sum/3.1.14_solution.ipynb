```python
# Solution : Bruteforce 

# compute length of the longest subarray with sum 0
def solve(a: list[int]) -> int:
    # store best length found so far
    max_len = 0
    # map prefix sum -> first index seen
    sum_index = {}
    # running prefix sum
    s = 0

    # iterate through the array
    for i, val in enumerate(a):
        # update running sum
        s += val

        # if sum is zero, subarray [0..i] has zero sum
        if s == 0:
            # update best length
            max_len = i + 1
        # if this sum seen before, subarray (prevIndex..i] has zero sum
        elif s in sum_index:
            # maximize length using previous index
            max_len = max(max_len, i - sum_index[s])
        # first time seeing this sum, store its index
        else:
            sum_index[s] = i

    # return best length
    return max_len

# program entry
def main():
    # sample input
    a = [9, -3, 3, -1, 6, -5]
    # print result
    print(solve(a))

# run main
if __name__ == "__main__":
    main()
```

```python
# Solution : Most optimal approach for subarrays . Considers both positive and negative numbers within the subarray.


def longest_zero_sum_subarray(arr: list[int]) -> int:
    prefix_sum = 0
    max_length = 0

    # Stores the first occurrence of each prefix sum
    first_occurrence = {}

    for index, value in enumerate(arr):
        prefix_sum += value

        # If prefix sum becomes 0, the subarray from
        # index 0 to current index has sum 0
        if prefix_sum == 0:
            max_length = index + 1

        # If we've seen this prefix sum before,
        # the elements between the previous index
        # and current index sum to 0
        elif prefix_sum in first_occurrence:
            current_length = index - first_occurrence[prefix_sum]
            max_length = max(max_length, current_length)

        # Store only the first occurrence to maximize length
        else:
            first_occurrence[prefix_sum] = index

    return max_length

arr = [6, -2, 2, -8, 1, 7, 4, -10]
print(longest_zero_sum_subarray(arr))
```

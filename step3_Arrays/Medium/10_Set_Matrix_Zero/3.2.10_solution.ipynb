# Problem: Set Matrix Zero


Problem Statement: Given a matrix if an element in the matrix is 0 then you will have to set its entire column and row to 0 and then return the matrix..
(See examples)

```python
# My solution :

def set_matrix_to_zero(mat):
    rows = len(mat)
    cols = len(mat[0])
    zero_cols = []
    print(mat)
    res = mat.copy()    # This is not a good approach to making copy of the file. This is not making a good deep copy.
    for i in range(rows):
        for j in range(cols):
            if mat[i][j] == 0:
                res[i] = [0]*cols
                zero_cols.append(j)

    for i in range(rows):
        for j in zero_cols:
            res[i][j] = 0    

    return res

arr = [[1,1,1],[1,0,1],[1,1,1]]
res_mat = set_matrix_to_zero(arr)
print(res_mat)
```

```python
# Solution : Bruteforce approach
# Here the matrix is iterated and if we find an element with a value of 0 we make the non zero elements in the coresponding matrix to -1.
# Only the non zero ones are choosed because if we convert every element to zero then we dont know which elements were actually zero so that their corresponding rows and more importantly columns can be transformed to zeros.
# Thats why we convert only the non -zero elements of the corresponing rows and columns into -1

class Solution:
     # Function to set entire row and column to 0 if an element in the matrix is 0
    def setZeroes(self, matrix):
        # Get number of rows
        m = len(matrix)
        # Get number of columns
        n = len(matrix[0])

        # First pass: mark rows and columns
        for i in range(m):
            for j in range(n):
                # If current cell is zero
                if matrix[i][j] == 0:
                    # Mark entire row
                    for col in range(n):
                        if matrix[i][col] != 0:
                            matrix[i][col] = -1
                    # Mark entire column
                    for row in range(m):
                        if matrix[row][j] != 0:
                            matrix[row][j] = -1

        # Second pass: replace -1 with 0
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == -1:
                    matrix[i][j] = 0

# Driver code
matrix = [[1,1,1],[1,0,1],[1,1,1]]
print(f"GIVEN : {matrix}")
sol = Solution()
sol.setZeroes(matrix)
for row in matrix:
    print(row)
```

```python
# Solution : Better approach
# Here we mark down the row and column values of the elements with value as zero while iterating the matrix.
# then we iterate again 

class Solution:
    # Function to set entire row and column to 0 if an element in the matrix is 0
    def setZeroes(self, matrix):
        # Get number of rows
        m = len(matrix)
        # Get number of columns
        n = len(matrix[0])

        # Create row marker array
        row = [0] * m
        # Create column marker array
        col = [0] * n

        # First pass: mark rows and columns that need to be zeroed
        for i in range(m):
            for j in range(n):
                # If element is zero, mark its row and column
                if matrix[i][j] == 0:
                    row[i] = 1
                    col[j] = 1

        # Second pass: set cells to zero based on markers
        for i in range(m):
            for j in range(n):
                # If the row or column is marked, set cell to zero
                if row[i] == 1 or col[j] == 1:
                    matrix[i][j] = 0

# Driver code
matrix = [[1,1,1],[1,0,1],[1,1,1]]
obj = Solution()
obj.setZeroes(matrix)
for r in matrix:
    print(r)
```

# Optimal approach:
# It is an extension of the aboue "better" approach
# Instead of storing seperate list for marking which all rows and columns needed to be made zero , we can store that value in the first row and column of the matrix
# we keep two variables (flags) to mark if the first row or column needed to transformed to zeros.

# Algorithm:
1. Check if the first row has any zero and store in a boolean flag.
2. Check if the first column has any zero and store in a boolean flag.
3. Traverse the rest of the matrix:
4. If a cell is zero, mark its row in the first column and its column in the first row as zero.
5. Traverse again (excluding first row and column), setting cells to zero if their row marker or column marker is zero.
6. Finally, update the first row and first column based on the stored flags.

```python
# Optimal approach:
class Solution:
    # Function to set entire row and column to 0 if an element in the matrix 
    def setZeroes(self, matrix):
        # Get dimensions of matrix
        m = len(matrix)
        n = len(matrix[0])

        # Flag to track if first row should be zeroed
        first_row_zero = False
        # Flag to track if first column should be zeroed
        first_col_zero = False

        # Check if first row has any zero
        for j in range(n):
            if matrix[0][j] == 0:
                first_row_zero = True
                break

        # Check if first column has any zero
        for i in range(m):
            if matrix[i][0] == 0:
                first_col_zero = True
                break

        # Use first row/column as markers
        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][j] == 0:
                    matrix[i][0] = 0
                    matrix[0][j] = 0

        # Set cells to zero based on markers
        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][0] == 0 or matrix[0][j] == 0:
                    matrix[i][j] = 0

        # Zero the first row if needed
        if first_row_zero:
            for j in range(n):
                matrix[0][j] = 0

        # Zero the first column if needed
        if first_col_zero:
            for i in range(m):
                matrix[i][0] = 0


# Driver code
matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
Solution().setZeroes(matrix)
for row in matrix:
    print(row)
```

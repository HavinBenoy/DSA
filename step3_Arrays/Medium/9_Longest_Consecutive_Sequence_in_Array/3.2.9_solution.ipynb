Problem : Longest Consecutive Sequence in an Array

Problem Statement: Given an array nums of n integers.

Return the length of the longest sequence of consecutive integers. The integers in this sequence can appear in any order.



```python
# My solution : This is a better approach. 
# Which is more optimal than the bruteforce approach in which we could check whether the existence of consecutive numbers of every element present in the array.
# Here the time complexity is O(nlogn), due to the sorting step.

def any_order_consecutive_max_count(arr):
    arr.sort()
    count = 1
    max_count = -1
    for i in range(1,len(arr)):
        if arr[i-1] != arr[i]:
            if arr[i-1]+1 == arr[i]:
                count +=1
                max_count = max(max_count,count)
            else:
                count = 0
    return max_count

arr = [100, 4, 200, 1, 3, 2,100]
maxi_count = any_order_consecutive_max_count(arr)
print(f"The maximum lenth of the consecutive elements in the given array is {maxi_count}")
```

# optimal apprach 
we use the set data structure to store the values of array.

This is because of the reason that , set is an unordered collection of unique element simplemented using a hash table

so it does have a very fast membership testing feature. And we can directly check it with the value of the arr elements itself.

Also :
The time complexity is O(n) on average, even though there is a while loop nested inside a for loop.

This is because if we trace the solution , we can identify that only 1 time the array got executed.
All the other times, when the bigger elements of the sequence appeared they were instantly skipped from entering the while loop.

```python
#Optimal approach

class Solution:
    def longestConsecutive(self, nums):
        n = len(nums)
        # If the array is empty
        if n == 0:
            return 0 

        # Initialize the longest sequence length
        longest = 1 
        st = set()

        # Put all the array elements into the set
        for i in range(n):
            st.add(nums[i])

        # Traverse the set to find the longest sequence
        for it in st:
            # Check if 'it' is a starting number of a sequence
            if it - 1 not in st:
                # Initialize the count of the current sequence
                cnt = 1 
                # Starting element of the sequence
                x = it 

                # Find consecutive numbers in the set
                while x + 1 in st:
                    # Move to the next element in the sequence
                    x = x + 1 
                    # Increment the count of the sequence
                    cnt = cnt + 1 
                # Update the longest sequence length
                longest = max(longest, cnt)
        return longest

if __name__ == "__main__":
    a = [100, 4, 200, 1, 3, 2] 
    # Create an instance of the solution class
    solution = Solution() 
    # Function call to find the longest consecutive sequence
    ans = solution.longestConsecutive(a) 
    print("The longest consecutive sequence is", ans)
```

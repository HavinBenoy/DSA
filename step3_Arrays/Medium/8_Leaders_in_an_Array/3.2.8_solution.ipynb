Problem : Leaders in an Array

Problem Statement:identify and return all elements in a given array that are strictly greater than (or equal to) all elements on their right side. By definition, the rightmost element in the array is always considered a leader since there are no elements to its right.

Example 1:
Input:
 arr = [4, 7, 1, 0]  
Output:
 7 1 0  
Explanation:
 The rightmost element (0) is always a leader.  
7 and 1 are greater than the elements to their right, making them leaders as well.

Example 2:
Input:
 arr = [10, 22, 12, 3, 0, 6]  
Output:
 22 12 6  
Explanation:
 6 is a leader because there are no elements after it.  
12 is greater than all the elements to its right (3, 0, 6), and 22 is greater than 12, 3, 0, 6, making them leaders as well.

```python
# My solution : THis is the most optimal solution rather than a bruteforce approach with inner and outer loops to check all the combinations.

def leaders_in_an_array(arr):
    
    leaders = []
    
    if not arr:
        return 
    n = len(arr)
    
    right_max_elem = arr[n-1]
    for i in range(n-1,-1,-1):
        if arr[i] >= right_max_elem:
            leaders.append(arr[i])
            right_max_elem = arr[i]
    return leaders

arr = [10, 22, 12, 3, 0, 6]
leaders = leaders_in_an_array(arr)
print(f"The leaders in the array are : {leaders[::-1]}") 
```
